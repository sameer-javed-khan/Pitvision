# PitVision — Notebook 1: Exploratory Data Analysis (EDA)


Goal: actually *look* at the cleaned data before training anything. EDA helps you
catch problems early (weird values, missing compounds) and builds intuition about
how tires degrade.




```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_parquet("../data/features/features.parquet")
print("Rows:", len(df))
print("Columns:", list(df.columns))
df.head()
```

## 1. Basic sanity checks

Always start here. We confirm there are no missing values and look at the
spread of lap times.

```python
print(df.isna().sum())


print("\nLap time (seconds) summary:")
print(df["LapTimeSeconds"].describe())
```

## 2. Distribution of lap times

A histogram shows the range of lap times. Most laps should cluster in a sensible
band; a long tail of very slow laps would suggest cleaning missed something.

```python
plt.figure(figsize=(8, 4))
plt.hist(df["LapTimeSeconds"], bins=50)
plt.xlabel("Lap Time (seconds)")
plt.ylabel("Number of laps")
plt.title("Distribution of Lap Times")
plt.show()
```

## 3. The key relationship: tire age vs lap time

This is the heart of the project. As a tire gets older (higher TyreLife), lap
times should generally get *slower*. If you see an upward trend here, your data
captures real degradation.

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["TyreLife"], df["LapTimeSeconds"], alpha=0.2, s=10)
plt.xlabel("Tire Age (laps)")
plt.ylabel("Lap Time (seconds)")
plt.title("Tire Age vs Lap Time (all races combined)")
plt.show()
```

## 4. Compound comparison

Different compounds degrade differently. SOFT tires are usually fastest but wear
out quickest; HARD tires are slower but last longer. Let's check average lap time
per compound.

```python
def compound_label(row):
    if row["compound_SOFT"] == 1: return "SOFT"
    if row["compound_MEDIUM"] == 1: return "MEDIUM"
    if row["compound_HARD"] == 1: return "HARD"
    return "UNKNOWN"

df["CompoundLabel"] = df.apply(compound_label, axis=1)
print(df.groupby("CompoundLabel")["LapTimeSeconds"].mean())
```
